# Filtering

```python
# Making a salary above certain amount

df

```

```python
countries = ['Finland','India']
```

```python
high_salary = ((df['ConvertedComp'] > 100000) & (df['Country'].isin(countries)))
#For checking for multiple values
```

```python
df.loc[high_salary,['Country' , 'LanguageWorkedWith' , 'ConvertedComp']]
```

```python
# To check if the person having a salary greator than 100000 have python

newcondition = df['LanguageWorkedWith'].str.contains('Python' , na = False)
   
new_filt = ((df['ConvertedComp'] > 100000) & (df['Country'].isin(countries)) & newcondition)
```

```python
df.loc[new_filt,['Country' , 'LanguageWorkedWith' , 'ConvertedComp']]
```
